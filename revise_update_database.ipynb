```python
import pandas as pd
```

```python
%%time
db = pd.read_csv('From100G.csv',usecols=['tpep_picku']) #import pickup time sonly
df = pd.to_datetime(db['tpep_picku']) #convert the column to timestamp
df = df.dt.round('1min') #round the times to the nearest minute, aka drop seconds
df = df.reset_index(name = 'Time') #label the column Time
```
